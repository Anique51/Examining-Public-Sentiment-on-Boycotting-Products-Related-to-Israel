## 1. Plotting Income Levels on a Pie Chart ##

```python
income_counts = df['Income Level'].value_counts()
```

```python
income_counts
```

```python
plt.figure(figsize=(10, 5))
plt.pie(income_counts,autopct='%1.1f%%', startangle=140)
plt.title('Income Levels Distribution')
plt.axis('equal')
plt.legend(loc='upper left', labels=income_counts.index)
plt.tight_layout()
plt.axis('off')
plt.savefig('Figure 1.png')
```

## 2. Plotting Social Media Usage on a Bar Chart ##

```python
social_mediaUsage_count = df['Social Media Usage'].value_counts()
```

```python
social_mediaUsage_count
```

```python
plt.figure(figsize=(10,6))
social_mediaUsage_count.plot(kind='bar', color='skyblue')
plt.title("Distribution of Social Media Usage Per Day")
plt.xlabel("Social Media Usage")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Figure 2.png')
```

## 3. Plotting Previous Activism Participation on Count Plot ##

```python
plt.figure(figsize=(8, 6))  # Set the figure size
order = df['Previous Activism Participation'].value_counts().index
sns.countplot(y=df['Previous Activism Participation'], palette='Set2', order = order)
plt.title('Distribution of Previous Activism Participation')
plt.ylabel('Levels of Participation')
plt.xlabel('Frequency')
plt.tight_layout()

plt.savefig('Figure 3.png')  # Save the figure
plt.show()
```
